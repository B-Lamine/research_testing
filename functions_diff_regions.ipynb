```python
import numpy as np
from random import random
from math import *
from sympy import *
from fractions import Fraction
from decimal import Decimal,getcontext
import random
import math
import cmath
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb

def F(m,n):
 if m == n:
  return 0
 else:
  return ((m+n)/2-m*n/(m-n)*math.log(m/n)) 

def FF(a,b,ra,rb) :
  mj2=a*complex(1,-ra)
  ml2=b*complex(1,-rb)
  if ra == 0. and rb == 0. :
    return F(a,b)
  elif mj2 == ml2 :
    return 0.0
  else :  
    return ((mj2+ml2)/2-mj2*ml2*cmath.log(mj2/ml2)/(mj2-ml2)).real

#x = np.random.uniform(0,3000,10000)
#y = np.random.uniform(0,3000,10000)
r = [(0.0001,0.0),(0.001,0.0),(0.01,0.0),(0.05,0.0),(0.1,0.0),(0.2,0.0),(0.05,0.1),(0.1,0.2),(0.18,0.2)]

#f = open("FF_py","w")
#for a in x:
# for b in y:
#  f.write(str(format(a,"10.4E"))+' '+str(format(b,"10.4E")) \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.0001,0.)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.001,0.)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.01,0.)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.05,0.)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.1,0.)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.2,0.)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.05,0.1)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.1,0.2)),"10.4E") \
#          +' '+str(format(F(a**2,b**2)-FF(a**2,b**2,0.18,0.2)),"10.4E")+'\n')
#f.close()
#sys.exit(0)


fig, axes = plt.subplots(3,3, figsize=(15,15))
norm=matplotlib.colors.LogNorm(vmin=10**(-4), vmax=10**4)

for ax,(ra,rb) in zip(axes.flat,r):
  xx=[]
  yy=[]
  delf=[]
  for i in range(100):
    a=(10*300**(i/100))
    for k in range(100):
      b=(10*300**(k/100))
      xx.append(a)
      yy.append(b)
      delf.append(F(a**2,b**2)-FF(a**2,b**2,ra,rb))
  delF=np.asarray(delf)
  scat = ax.scatter(xx,yy,c=delF, cmap='CMRmap_r',norm=norm)
  ax.set_xscale("log")
  ax.set_xlim(10,3000)
  ax.set_yscale("log")
  ax.set_ylim(10,3000)
plt.colorbar(scat, ax=axes.ravel().tolist())
plt.show();
```
